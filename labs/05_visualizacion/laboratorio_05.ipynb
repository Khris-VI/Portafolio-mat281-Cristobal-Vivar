```python
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(rc={'figure.figsize':(10,8)})
```

```python
# cargar datos
df = pd.read_csv(os.path.join("data","company_sales_data.csv"))
df.head()
```

El objetivo es tratar de obtener la mayor información posible de este conjunto de datos. Para cumplir este objetivo debe resolver las siguientes problemáticas:

1. Lea el "total_profit" de todos los meses, muéstrelo usando un gráfico lineal y un gráfico de dispersión.

```python
# tamano del grafico
fig = plt.figure(figsize=(15, 6)) # ventana

plt.subplot(1, 2, 1) # sub-ventana
plt.plot(
    df['month_number'], # eje x
    df['total_profit'], # eje y
    label='total_profit', # etiquetado
    color="blue", # color
    linewidth=1 # tamano de la curva
)

plt.legend() # agregar etiquetado
plt.title("Gráfico lineal") # agregar titulo
plt.xlabel('month_number') # nombre eje x
plt.ylabel('total_profit') # nombre eje y
plt.xticks(df['month_number'])
plt.grid() # agregar grillado

plt.subplot(1, 2, 2) # sub-ventana
plt.scatter(
    df['month_number'], # eje x
    df['total_profit'], # eje y
    s = 100, #Tamaño de los puntos
    c="red", # color
    alpha=1, # intensidad color
    marker='o', # forma de los puntos
    label='total_profit' # etiquetdo fijando posicion
)

plt.legend() # agregar etiquetado
plt.title('Gráfico de Dispersión') # agregar titulo
plt.xlabel('month_number') # nombre eje x
plt.ylabel('total_profit') # nombre eje y
plt.xticks(df['month_number']) #Forzado de numeración total de los meses
plt.grid() # agregar grillado

fig.tight_layout(pad=6.0) #Separación entre gráficos
plt.show() # mostrar grafico
```

2. Lea todos los datos de ventas de productos y muéstrelos utilizando un gráfico multilínea.

```python
# tamano del grafico
fig = plt.figure(figsize=(15, 10)) # ventana

palette = [ #Colores
    '#00FF11',  # Facecream
    '#E6FF00',  # Facewash
    '#6890F0',  # Toothpaste
    '#9A00FF',  # Bathingsoap
    '#FF0000',  # Shampoo
    '#000000',  # Moisturizer
]

for i in range(1,6): #Creación de los 5 primeros gráficos
    plt.plot(
        df['month_number'], # eje x
        df[df.columns[i]], # eje y
        label=df.columns[i], # etiquetado
        color=palette[i-1], # color
        linewidth=2 # tamano de la curva
    )
plt.plot( #Creación del último gráfico para diferenciarlo de facewash
    df['month_number'], # eje x
    df[df.columns[6]], # eje y
    label=df.columns[6], # etiquetado
    color=palette[5], # color
    linestyle='dotted', #Estilo de línea
    linewidth=3 # tamano de la curva
)

plt.xlabel('month_number',fontsize=18) # nombre eje x
plt.ylabel('Cantidad Vendida',fontsize=18) # nombre eje y
plt.xticks(df['month_number']) #Forzado de numeración total de los meses
plt.legend() # agregar etiquetado

plt.show()
```

3. Lea los datos de ventas de productos de "facecream" y "facewash" y muéstrelos usando el gráfico de barras.

```python
# tamano del grafico
fig = plt.figure(figsize=(15, 5)) # ventana

plt.subplot(1, 2, 1) # sub-ventana
plt.bar(
    df['month_number'], # eje x
    df['facecream'], # eje y 
    align='center', # centrar nombre eje x
    color='blue', # color 
    alpha=1 # intensidad del color
)
plt.xlabel('month_number',fontsize=18) # nombre eje x
plt.ylabel('facecream',fontsize=18) # nombre eje y
plt.xticks(df['month_number'])

plt.subplot(1, 2, 2) # sub-ventana
plt.bar(
    df['month_number'], # eje x
    df['facewash'], # eje y 
    align='center', # centrar nombre eje x
    color='red', # color 
    alpha=1 # intensidad del color
)
plt.xlabel('month_number',fontsize=18) # nombre eje x
plt.ylabel('facewash',fontsize=18) # nombre eje y
plt.xticks(df['month_number']) #Forzado de numeración total de los meses

fig.tight_layout(pad=6.0) #Separación entre gráficos
plt.show()
```

4. Lea todos los datos de ventas de productos y muéstrelos utilizando un gráfico box-plot.

```python
# Definición de dataframe con solo las columnas de interés
products_df = df.drop(['month_number','total_profit','total_units'], axis=1)
 
# Creación del Box-Plot
sns.set(font_scale = 1.5) #incremento del tamaño de letras
sns.boxplot(data=products_df).set(ylabel='Cantidad Vendida')

plt.show()
```

5. Calcule los datos de ventas totales del año pasado para cada producto y muéstrelos usando un gráfico circular

```python
ventas_totales={} #Creación de diccionario donde almacenar las ventas totales de cda producto

for i in range (1,7):
    ventas_totales[df.columns[i]]=df[df.columns[i]].sum() #Creación de cada llave
    
df_total=pd.Series(ventas_totales) #Conversión del diccionario a Serie
plt.pie( #Creación del gráfico circular
    df_total, # Tamaño de las secciones circulares
    labels = df.drop(['month_number','total_profit','total_units'], axis=1).columns, #Etiquetado
    shadow = True, #Sombra del gráfico
    autopct='%1.1f%%' #visualización de porcentaje
)

plt.title('Distribución de ventas totales del año pasado',fontsize=18) #Titulo del gráfico

plt.show()
```
